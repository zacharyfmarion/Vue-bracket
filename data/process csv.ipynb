```python
import pandas as pd
import json
```

```python
df = pd.read_csv('ncaa_results.csv')
```

```python
df
```

```python
def process_year(df, year):
    result = []
    year_df = df.loc[df['Year'] == year]
    for round_number in year_df.Round.unique():
        round_result = []
        round_df = year_df.loc[year_df['Round'] == round_number]
        for index, row in round_df.iterrows():
            round_result.append([
                {
                    "name": row['Team'],
                    "id": "{}.{}".format(row["Region Number"], row["Seed"]),
                    "winner": True 
                },
                {
                    "name": row['Team.1'],
                    # This is going to be incorrect if teams from different region meet
                    "id": "{}.{}".format(row["Region Number"], row["Seed"]),
                }
            ])
        result.append(round_result)
    return result
```

```python
process_year(df, 1985)
```

```python
def process_data(df, filepath):
    data = {}
    for year in range(1985, 2018):
        data[year] = process_year(df, year)
    with open(filepath, 'w') as outfile:
        json.dump(data, outfile, indent=4, sort_keys=True)
```

```python
process_data(df, 'data.json')
```
