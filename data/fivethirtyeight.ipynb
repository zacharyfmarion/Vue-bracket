```python
df = pd.read_csv('fivethirtyeight.csv')
```

```python
def get_teams():
    res = {}
    for region in ["East", "South", "Midwest", "West"]:
        region_df = df.loc[df["team_region"] == region]
        region_df = region_df.drop_duplicates(subset='team_name', keep="last")
        teams = []
        for index, team in region_df.iterrows():
            try:
                seed = int(team["team_seed"])
            except:
                seed = int(team["team_seed"][:-1])
            teams.append({
                "name": team["team_name"],
                "seed": seed
            })
        teams.sort(key=lambda k: k["seed"])
        res[region] = teams
    return res
```
